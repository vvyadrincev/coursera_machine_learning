```python
import numpy as np
import re
import pandas
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn import datasets


import sys
sys.path.append("../")
from utils import save_answ, logging
```

```python
data = datasets.load_boston()
X = data.data
X = preprocessing.scale(X)
y = data.target

print type(X), X.shape, type(y), y.shape
```

```python
def calculate_metric (X, y):
    kf = KFold(n_splits=5, random_state=42, shuffle=True)
    
    mean_squared_mean_max = -10000.
    p_max = 0
    
    for p in np.linspace(start=1., stop=10., num=200):
        regressor = KNeighborsRegressor(n_neighbors=5, weights='distance', p = p)
        scores = cross_val_score(X=X,y=y, cv=kf, n_jobs=-1, estimator=regressor, scoring='neg_mean_squared_error')
        mean_squared_mean = scores.mean()
        
        if mean_squared_mean > mean_squared_mean_max:
            mean_squared_mean_max = mean_squared_mean
            p_max = p

        logging.info(" p: %0.5f, mean_squared: %0.2f (+/- %0.2f)" % (p, mean_squared_mean, scores.std() * 2))
    
    print "MAX: ", mean_squared_mean_max, p_max
    return p_max, mean_squared_mean_max

p_max, mean_squared_mean_max = calculate_metric (X, y)

save_answ(number=1, answer=("%0.1f" % p_max) )
```
